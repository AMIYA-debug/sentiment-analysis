```python
import numpy as np
import pandas as pd
data = pd.read_csv('all_kindle_review.csv')
data
```

```python
df=data[['reviewText','rating']]
df
```

```python
df.isnull().sum()
```

```python
df['rating']=df['rating'].apply(lambda x:1 if x>3 else 0 )
df['reviewText']=df['reviewText'].str.lower()
```

```python
df['rating'].value_counts()
```

```python
df
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wl=WordNetLemmatizer()
```

```python
def cleaning(sentence):
   review=re.sub('[^a-zA-z]',' ',sentence)
   review=review.lower()
   review=review.split()
   return review
def processing(sentence):
   words=[]
   for i in range(0,len(sentence)):
      if sentence[i] not in stopwords.words('english'):
          sentence[i]=wl.lemmatize(sentence[i],pos='v')
          words.append(sentence[i])

   return words
df['reviewText']=df['reviewText'].apply(lambda x: processing(cleaning(x)))

```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df['reviewText'],df['rating'],random_state=42,test_size=0.25,)
x_train
```

```python
from gensim.models import keyedvectors
model = "C:/Users/amiya/OneDrive/Documents/word2vecmodel/archive/GoogleNews-vectors-negative300.bin"
vectoriser = keyedvectors.load_word2vec_format(model, binary=True)
```

```python
def avgw2v(doc):
  return np.mean([vectoriser[word] for word in doc if word in vectoriser.key_to_index],axis=0) 
```

```python
from tqdm import tqdm
x_train_b = []
for i in tqdm(range(len(x_train))):
    x_train_b.append(avgw2v(x_train.iloc[i]))
x_train_b=np.array(x_train_b)
x_train_b
```

```python
y_train=np.array(y_train)
x_train_b.shape
```

```python
train_data=pd.DataFrame(x_train_b)
train_data['output']=y_train
train_data
```

```python
from sklearn.naive_bayes import GaussianNB
ML_model=GaussianNB()
x = train_data.iloc[:,:-1]
y=train_data.iloc[:,-1]
ML_model.fit(x, y)
```

```python
x_test_b = []
for i in tqdm(range(len(x_test))):
    x_test_b.append(avgw2v(x_test.iloc[i]))
x_test_b=np.array(x_test_b)
x_test_b
```

```python
y_test=np.array(y_test)
x_test_b.shape
```

```python
test_data=pd.DataFrame(x_test_b)
test_data['output']=y_test
test_data
```

```python
x_test=test_data.iloc[:,:-1]
y_test=test_data.iloc[:,-1]
y_predict=ML_model.predict(x_test)
y_predict
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
score=accuracy_score(y_test,y_predict)
score
```

```python
classification_report(y_test,y_predict)
```

```python
confusion_matrix(y_test,y_predict)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(n_jobs=-1,random_state=42)
lr.fit(x,y)
```

```python
y_predict_2=lr.predict(x_test)
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
score=accuracy_score(y_test,y_predict_2)
score
```

```python
classification_report(y_test,y_predict_2)
```

```python
confusion_matrix(y_test,y_predict_2)
```
